## Binomial Distribution Example

### Scenario

Suppose you have a biased coin that lands on heads 70% of the time (\(p = 0.7\)) and tails 30% of the time (\(q = 0.3\)). You decide to flip this coin 10 times and want to determine the probability of getting exactly 7 heads.

### Parameters
- **Number of trials (\(n\))**: 10
- **Probability of success (\(p\))**: 0.7 (probability of getting heads)
- **Probability of failure (\(q\))**: 0.3 (probability of getting tails)
- **Number of successes (\(k\))**: 7

### Binomial Probability Formula

The probability of getting exactly \(k\) successes in \(n\) independent Bernoulli trials is given by the binomial distribution formula:

\[ P(X = k) = \binom{n}{k} p^k q^{n-k} \]

where:
- \( \binom{n}{k} \) is the binomial coefficient, calculated as \( \frac{n!}{k!(n-k)!} \)
- \( p \) is the probability of success on an individual trial
- \( q \) is the probability of failure on an individual trial (\( q = 1 - p \))
- \( n \) is the number of trials
- \( k \) is the number of successes

### Calculation

1. **Calculate the binomial coefficient \( \binom{n}{k} \)**:
   \[
   \binom{10}{7} = \frac{10!}{7!(10-7)!} = \frac{10!}{7! \cdot 3!}
   \]
   
2. **Substitute the values into the formula**:
   \[
   P(X = 7) = \binom{10}{7} \times (0.7)^7 \times (0.3)^{10-7}
   \]

Let's compute this step-by-step.

### Step-by-Step Calculation

#### Binomial Coefficient
\[
\binom{10}{7} = \frac{10!}{7! \cdot 3!} = \frac{10 \times 9 \times 8}{3 \times 2 \times 1} = 120
\]

#### Probability Calculation
\[
P(X = 7) = 120 \times (0.7)^7 \times (0.3)^3
\]

\[
(0.7)^7 \approx 0.0823543
\]

\[
(0.3)^3 = 0.027
\]

\[
P(X = 7) = 120 \times 0.0823543 \times 0.027 \approx 0.2668
\]

Therefore, the probability of getting exactly 7 heads in 10 flips of this biased coin is approximately 0.2668, or 26.68%.

### Summary
The binomial distribution helps us calculate the probability of a specific number of successes in a fixed number of trials, given the probability of success on each trial. In this example, it helped us find the probability of getting exactly 7 heads out of 10 flips of a biased coin.
